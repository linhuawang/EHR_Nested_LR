```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("procedure.csv")
df.iloc[:3,:]
```

```python
df = df[["person_id","procedure_concept_id"]]
df["value"] = 1
df.iloc[:3,:]
```

```python
df.fillna(0, inplace=True)
df.procedure_concept_id = df.procedure_concept_id.astype(int)
```

```python
procedure_matrix = pd.pivot_table(data=df, columns="procedure_concept_id",  
                                  index="person_id",values="value", aggfunc=np.sum)
```

```python
del procedure_matrix.columns.name
del procedure_matrix.index.name
procedure_matrix.fillna(0, inplace=True)
```

```python
procedure_matrix = procedure_matrix.iloc[:,1:]
procedure_matrix
```

```python
person_table = pd.read_csv("person_final.csv",index_col=0)
person_table
```

```python
procedure_matrix["death"]=person_table["death"]
procedure_matrix.iloc[:3,:]
```

```python
cols = procedure_matrix.columns
cols = ["procedure_%d" %i for i in cols[:-1]] + [cols[-1]]
procedure_matrix.columns = cols
procedure_matrix
```

```python
procedure_matrix.to_csv("procedure_input.csv")
```
