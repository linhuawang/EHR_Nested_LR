```python
import pandas as pd
import numpy as np 
```

```python
drug = pd.read_csv("drug_ori.csv")
drug.iloc[:3,:]
```

```python
drug["total_dose"] = drug.quantity * drug.days_supply
drug = drug[["person_id","total_dose","drug_concept_id"]]
drug.iloc[:10,:]
```

```python
drug.fillna(0,inplace=True)
drug.drug_concept_id = drug.drug_concept_id.astype(int)
drug_matrix = pd.pivot_table(data=drug,index="person_id",columns="drug_concept_id", values="total_dose")
del drug_matrix.columns.name
del drug_matrix.index.name
drug_matrix
```

```python
drug_matrix.fillna(0,inplace=True)
drug_matrix.drop(columns=[0], inplace=True)
```

```python
person_matrix = pd.read_csv("person_input.csv", index_col=0)
person_matrix.iloc[:4,:]
```

```python
(drug_matrix.index == person_matrix.index).all()
```

```python
drug_matrix["death"] = person_matrix["death"]
```

```python
cols = drug_matrix.columns
cols = ["drug_%s" %i for i in cols]
drug_matrix.columns = cols
drug_matrix.to_csv("drug_input.csv")
```
